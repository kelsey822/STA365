## Part C (II): Full Conditional [Posterior] Distributions

### 1. What is $p(\theta|\tau, \mathbf{x})$?<br>This is a so-called a *full conditional* [posterior distribution]

#### *This is actually just asking for $p(\theta|\tau, \mathbf{x}) \propto p(\textbf{x}, \theta|\tau) \propto p(\textbf{x},\theta,\tau)$*<br>which is just $p(\textbf{x},\theta,\tau)$ viewed only as a function of $\theta$ with $\textbf{x}$ given<br>*(which means $\textbf{x}$ and $\tau$ are known because they've been observed)*

### 2. What is $p(\theta,\tau | \mathbf{x})$ this if all normalizing constants are removed if ONLY $\tau$ is was a random variable? 

#### *This is actually just asking for $p(\tau | \theta, \mathbf{x}) \propto p(\textbf{x},\tau |\theta) \propto p(x,\theta,\tau)$*<br>which is just $p(\textbf{x},\theta,\tau)$ viewed only as a function of $\tau$ with $\textbf{x}$ given<br>*(which means $\textbf{x}$ and $\theta$ are known because they've been observed)*

### That is, what is the *full conditional* [posterior distribution] $p(\tau|\theta, \mathbf{x})$?

$\LARGE
\begin{align*}
p(\tau|\theta, \mathbf{x}) &\propto{} f(\mathbf{x}|\theta)p(\tau|\theta)=f(\mathbf{x}|\theta)p(\tau)\\
&\propto{} \prod_{i=1}^n\sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2} \times 1 \\
&\propto{} \tau^{\frac{n}{2}} e^{-\frac{\tau}{2}\sum_{i=1}^n(x_i-\theta)^2} 
\end{align*}
$


### 3. What is $p(\tau|\theta, \mathbf{x})$ if we still have $p(\tau) \propto 1$<br>but now $p(\theta)$ is $N(\theta_0, \tau^2_0)$?

## Part C (III): Gibbs Sampling

The canonical example of ***Gibbs sampling*** is the ***normal-gamma*** model and here's a simplified version with $p(\theta,\tau) \propto 1$

$$\begin{align*}
p\left(x|\theta,\tau=\frac{1}{\sigma^2}\right) = {} & \prod_{i=1}^n \sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2}\\
p(\theta,\tau) \propto {} & 1 \\
p(\theta,\tau| x) \propto p(\theta,\tau, x) \propto {} &  \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2} \\
p(\tau|\theta, x) = {} & \text{Gamma}\left(\alpha = \frac{n}{2}+1, \beta= \frac{1}{2} \sum_{i=1}^n (x_i-\theta)^2 \right)\\
p(\theta| \tau, x)  \propto {} & e^{-\frac{\tau}{2} \sum_{i=1}^n \theta^2 - 2\theta x_i} =  e^{-\frac{\tau}{2} \left(n\theta^2 - 2\theta \sum_{i=1}^n x_i\right)} = e^{-\frac{\tau n }{2} \left(\theta^2 - 2\theta \bar x\right)}\\
  \propto {} & e^{-\frac{n \tau}{2} (\theta - \bar x)^2} \\
p(\theta| \sigma^2 = \tau^{-1}, x) = {} & N\left(E[\theta] = \bar x, Var(\theta) = \frac{\sigma^2}{n}\right)
\end{align*}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C): # c for chains
    for g in range(1,G): # the gibbs sampler where g is the number of gibbs samples
        # https://en.wikipedia.org/wiki/Gamma_distribution
        tau[c,g] = stats.gamma(n/2+1, scale= 2/((x-theta[c,g])**2).sum()).rvs() #\
        theta[c,g] = stats.norm(x.mean(), scale = 1/(tau[c,g]*n)**0.5).rvs()
        #\
```

```python
# loc is mean and scale is standard deviation
n = 30
m = 0
s = 1

x = stats.norm(loc=m, scale=s).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples

theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization

for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        shape = n/2
        scale = 2 / np.sum((x - theta[c, g - 1])**2)
        tau[c,g] = stats.gamma(a=shape, scale=scale).rvs()
        
        #update theta
        theta[c, g] = stats.norm(loc=np.mean(x), scale=np.sqrt(1 / (tau[c, g] * n))).rvs()

#to provide a demonstration of bayesian inference
burn = 20 
posterior_theta = theta[:, burn:].flatten() 
posterior_tau = tau[:, burn:].flatten()    

# 95% credible intervals
theta_credible_interval = np.percentile(posterior_theta, [2.5, 97.5])
tau_credible_interval = np.percentile(posterior_tau, [2.5, 97.5])

print("95% Credible Interval for :", theta_credible_interval)
print("95% Credible Interval for :", tau_credible_interval)
                
```

## Q2

First find the likelihood for $p(\textbf{x}|\theta, \tau)$. Let \textbf{x}=$x_1,..,x_n)$ i.i.d from $N(\theta, \tau^{-1})$. We know that $p(\tau, \theta|\textbf{x})$ is proportional to the likelihood and prior. Thus we have
\begin{align*}
    p(\textbf{x}|\theta, \tau) &\propto \tau^{\frac{n}{2}}\text{exp}(-\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2) \\
    p(\theta) &\propto \text{exp}(-\frac{\tau_0}{2}(\theta-\theta_0)^2\\
    p(\tau) &\propto \tau^{\frac{\alpha}{2-1}}e^{\frac{-\lambda\tau}{2}}
\end{align*}
From this information we then have
\begin{align*}
    p(\tau, \theta|\textbf{x}) \propto \tau^{\frac{n}{2}}\text{exp}(-\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2)\text{exp}(-\frac{\tau_0}{2}(\theta-\theta_0)^2
\end{align*}
Now to find $p(\tau|\theta, \textbf{x})$:
\begin{align*}
    (\tau|\theta, \textbf{x})&\propto  \tau^{\frac{n}{2}}\text{exp}(-\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2)\tau^{\frac{\alpha}{2-1}}e^{\frac{-\lambda\tau}{2}}\\
    &\propto \tau^{\frac{n+\alpha}{2}-1}\text{exp}(-\frac{\tau}{2}(\sum_{i=1}^{2}(x_i-\theta)^2+\lambda))\\
    \Rightarrow p(\tau|\theta, \textbf{x}) &\sim \text{Gamma}(\frac{n+\alpha}{2}, \frac{2}{\sum_{i=1}^{2}(x_i-\theta)^2+\lambda)})
\end{align*}
We do the same for $p(\theta|\tau, \textbf{x}):$
\begin{align*}
    p(\theta|\tau, \textbf{x}) &\propto \text{exp}(-\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2)\text{exp}(-\frac{\tau_0}{2}(\theta-\theta_0)^2 
\end{align*}
For simplification we have:
\begin{align*}
    \sum_{i=1}^{n} (x_i - \theta)^2)&= \sum_{i=1}^n x_i^2 -2\theta\sum_{i=1}^n x_i+n\theta^2\\
    (\theta-\theta_0)^2 &= \theta^2-2\theta\theta_0+\theta_0^2
\end{align*}
Therefore
\begin{align*}
    p(\theta|\tau, \textbf{x}) &\propto \text{exp}(\frac{-1}{2}((\tau n+\tau_0)\theta^2-2\theta(\tau\sum_{i=1}^n x_i+\tau_0\theta_0)))\\
    \Rightarrow  p(\theta|\tau, \textbf{x}) &\sim N(
    \frac {\tau \sum_{i=1}^n x_i+\tau_0\theta_0}{\tau n+\tau_0}, 
    \frac{1}{\tau n +\tau_0})
\end{align*}


To address the second part of the problem:

From $p(\tau)$ we know that $\tau \sim \text{Gamma}(\frac{\alpha}{2}, \frac{2}{\lambda})$. Therefore here $\alpha$ is the shape parameter, meaning it gives insight to how strong the prior belief is. $\lambda$ is the rate parameter and impacts the mean of the prior. The same follows for $p(\tau|\textbf{x}, \theta)$ where the shape and rate parameters for the Gamma distribution are $\frac{n+\alpha}{2}, \frac{2}{\sum_{i=1}^{n}(x_i-\theta)^2+\lambda)}$, respectively.

$\alpha$ and $\lambda$ control the amount and nature of info the prior contributes to the posterior in this way. Since  $\frac{\alpha}{2}$ determines the number of observations contributed by the distributions. So the prior contrinbutes $\frac{\alpha}{2}$ observations and the likelihood will contribute $\frac{n}{2}$. With $\lambda$ as the rate parameter it impacts the mean of the prior and tells us about the variance of the prior 


```python
def gibbs_sampler(alpha, lmbda, theta_0, tau_0, n, x, C =2,G=1000, burn=20):

    theta,tau = np.zeros([C,G]),np.zeros([C,G])
    theta[:,0] = 1000  # initialization
    tau[:, 0] = 1 

    for c in range(C):
        for g in range(1,G):
            # https://en.wikipedia.org/wiki/Gamma_distribution
            shape = (n + alpha) /2
            scale = 1/(np.sum((x - theta[c, g - 1])**2) + lmbda) / 2
            tau[c,g] = stats.gamma(a=shape, scale=scale).rvs()
        
            #update theta
            theta[c, g] = stats.norm((tau[c, g] * np.sum(x) + tau_0 * theta_0) / (tau[c, g] * n + tau_0), scale=np.sqrt(1 / (tau[c, g] * n + tau_0))).rvs()

    #to provide a demonstration of bayesian inference
    posterior_theta = theta[:, burn:].flatten() 
    posterior_tau = tau[:, burn:].flatten()    

    # 95% credible intervals
    theta_credible_interval = np.percentile(posterior_theta, [2.5, 97.5])
    tau_credible_interval = np.percentile(posterior_tau, [2.5, 97.5])
    
    return posterior_theta, posterior_tau, theta_credible_interval, tau_credible_interval
```

```python
n = 30
m = 0
s = 1
x = stats.norm(loc=m, scale=s).rvs(size=n)

# try different priors to compare
prior_settings = [
    {"alpha": 6, "lmbda": 4, "theta_0": 0, "tau_0": 1},
    {"alpha": 10, "lmbda": 5, "theta_0": 0.2, "tau_0": 1},  
    {"alpha": 0.5, "lmbda": 0.5, "theta_0": 0, "tau_0": 1}, 
    {"alpha": 2, "lmbda": 2, "theta_0": 5, "tau_0": 1}, 
]

# gibbs sampler for each
results = []
for setting in prior_settings:
    result = gibbs_sampler(**setting, n=n, x=x)
    results.append((setting, *result))  

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
for i, (setting, posterior_theta, posterior_tau, theta_ci, tau_ci) in enumerate(results):
    row, col = divmod(i, 2)  
    ax[row, col].hist(posterior_theta, bins=50, density=True, alpha=0.5, color='blue', label=f"$\\theta$: ={setting['alpha']}, ={setting['lmbda']}")
    ax[row, col].hist(posterior_tau, bins=50, density=True, alpha=0.5, color='green', label=f"$\\tau$: ={setting['alpha']}, ={setting['lmbda']}")
    ax[row, col].axvline(theta_ci[0], color='red', linestyle='--', alpha=0.7, label=' CI: 2.5%')
    ax[row, col].axvline(theta_ci[1], color='red', linestyle='--', alpha=0.7, label=' CI: 97.5%')
    ax[row, col].axvline(tau_ci[0], color='orange', linestyle='--', alpha=0.7, label=' CI: 2.5%')
    ax[row, col].axvline(tau_ci[1], color='orange', linestyle='--', alpha=0.7, label=' CI: 97.5%')
    ax[row, col].set_title(f"Prior ={setting['alpha']}, ={setting['lmbda']}")

for a in ax.flatten():
    a.legend()

plt.tight_layout()
plt.show()
```

Once we plug this back in to Bayes formula and cancel and rearrange terms, this amounts to multiplying the normalizing constant of the posterior with the portion of the posterior that relies on the parameter. This sort of derivation isn't unique to conjugate priors; as long as we can recognize the family in which the product of the prior and likelihood belong to, the integral in the denominator can be solved by simplifying the term in the integral to the precise density (which occurs by moving normalizing constants outside of the integral, and introducing the normalizing constant for that new density). This is the technique we used to solve the integral above.

As a result, we can use a proportionality argument to skip all of these mathematical derivations. If we can recognize the functional form of the product between the likelihood and prior as being in the family of another probability distribution, then that is equivalent to moving around these normalizing constants to simplify integrals.
